```python
for index, row in matches_df.iterrows():
    player1, player2, date = row['Player 1'], row['Player 2'], row['Date']
    subset = matches_df[((matches_df['Player 1'] == player1) & (matches_df['Player 2'] == player2)) | ((matches_df['Player 1'] == player2) & (matches_df['Player 2'] == player1))]
    subset = subset[subset.Date < date]
    prev_matches = len(subset)
    p1_wins = len(subset[subset.Winner == player1])
    if prev_matches == 0:
        matches_df.at[index, 'Player 1 H2H Win Percentage'] = 0.5
        continue
    percent_p1 = p1_wins/prev_matches
    matches_df.at[index, 'Player 1 H2H Win Percentage'] = percent_p1
    print(index)
```

```python
def add_surfaces(tournaments_df):
    def custom_selector(tag):
        if tag.name == 'div' and tag.has_attr('class') and 'item-value' in tag.get('class'):
            return True
        return False
    surfaces = []
    i = 0
    for index in tournaments_df.index:
        name, id = index[0], tournaments_df.loc[index, 'Id'][0]
        url = 'https://www.atptour.com/en/tournaments/%s/%d/overview' % (name, id)
        t_page = requests.get(url).text
        soup = BeautifulSoup(t_page)
        surface = ""
        for tag in soup.find_all(custom_selector):
            tag = str(tag)
            content = tag[tag.find('>') + 1:tag.find('</div')]
            if content in ['Hard', 'Clay', 'Grass']:
                surface = content
        if surface == "":
            print(index)
        surfaces.append(surface)
        print(i, surface)
        i+=1
    tournaments_df['Surface'] = surfaces
```
